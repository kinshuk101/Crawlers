```python
df=pd.read_excel("data_set_wool.xlsx")
```

```python
df
```

```python
df.dropna(subset='EAN',inplace=True)
df.drop_duplicates(subset='EAN',inplace=True)
df.dropna(subset='URL',inplace=True)
df.drop_duplicates(subset='URL',inplace=True)
df.reset_index(drop=True, inplace=True)
df['seller_info']=''
df.shape #this will give you number of rows and coloumns
```

```python
for i in tqdm(range(0,len(df['URL']))):
    url=df['URL'][i]
    
    print('************************')
    print(i)
    print(url)

    driver.get(url)
    bs=BeautifulSoup(driver.page_source, "html.parser")
    try:
        div=bs.find("div",class_="partner")
        seller=div.find('a')
        df['seller_info'][i]=seller.text
        print(seller.text)
    except:
        print('woolworths')
```
