```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")
driver = webdriver.Chrome(executable_path = r'chromedriver.exe',chrome_options = chrome_options)
```

```python
for i in tqdm(range(0,5)):
    Url='https://www.liverpool.com.mx/tienda/pdp/-/'+df['sku'][i]+'?skuid='+df['sku'][i]
#     chrome_options = webdriver.ChromeOptions()
#     chrome_options.add_argument("--incognito")
#     driver = webdriver.Chrome(executable_path = r'chromedriver.exe',chrome_options = chrome_options)

    print('************************')
    print(i)
    print(Url)

    driver.get(Url)
    time.sleep(2)
    bs=BeautifulSoup(driver.page_source, "html.parser")
    
    try:
        aa=bs.find('li',class_="active")
        
        cat=[aa.text.strip()]
        print("category:",cat)
           
        df['categories'][i]=cat
        
    except:
        
        df['categories'][i]=""
        
        
    try:
        seller=bs.find('a', class_="a-productInfo__selledByLink").text
        print("sold by:" ,seller)
        df['sold_by'][i]=seller
    except:
        df['sold_by'][i]=""
        
    try:
        rating=bs.find('div',class_="RRBox").text
        print("rating_count:" ,rating)
        df['rating_count'][i]=rating
    except:
        df['rating_count'][i]=""    
        
```
